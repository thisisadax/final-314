```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import fractional_matrix_power
import networkx as nx
```

```python
'''
Compute and return a mask of edges in the MST of a given graph described by the input adjacency matrix.
'''
def mst_edges(am):
    am[np.isnan(am)] = 0
    am[am<0] = 0
    am_inv = 1 / am
    G = nx.convert_matrix.from_numpy_array(am_inv)
    mst = nx.algorithms.tree.mst.minimum_spanning_tree(G)
    mst_am = nx.to_numpy_array(mst)
    mst_am = (mst_am + mst_am.T) / 2
    mst_mask = mst_am>0
    return mst_mask

'''
Threshold the inputs adjacency matrix to a threshold level of sparsity (row or global) while ensuring graph connectedness using the graph's MST.
'''
def threshold(am, thresh, mode='row'):
    if mode=='row':
        thresh_mask = np.apply_along_axis(lambda x: x>np.quantile(x, thresh), 0, am)
    else:
        thresh_val = np.quantile(am.ravel(), thresh)
        thresh_mask = am>thresh_val
    mst_mask = mst_edges(am)
    # Threshold while ensuring graph connectedness.
    mask = thresh_mask | mst_mask
    am[~mask] = 0
    am = am + am.T - np.diag(am.diagonal())
    return am

'''
Normalize the input adjacency matrix.
'''
def normalize(am):
    #n = am.shape[0]
    #D = np.zeros((n, n))
    #deg = am.sum(0)
    #np.fill_diagonal(D, deg)
    #P = fractional_matrix_power(D, -0.5) @ am @ fractional_matrix_power(D, -0.5)
    #assert np.all_close(P.sum(0)==1)
    #np.fill_diagonal(P, 0)

    # Ensure matrix is PSD
    eigenvalues, eigenvectors = np.linalg.eigh(am)
    eigenvalues[eigenvalues < 0] = 0  # Set negative eigenvalues to zero
    am_psd = eigenvectors @ np.diag(eigenvalues) @ eigenvectors.T
    #row_sums = am_psd.sum(axis=1, keepdims=True)
    #row_sums[row_sums == 0] = 1
    #P = am_psd / row_sums
    #####
    #n = am.shape[0]
    #D = np.zeros((n, n))
    #deg = am.sum(0)
    #np.fill_diagonal(D, deg)
    #P = fractional_matrix_power(D, -0.5) @ am_psd @ fractional_matrix_power(D, -0.5)
    #np.fill_diagonal(P, 0)
    return am_psd

'''
Compute the diffusion embedding for the input adjacency matrix.
'''
def embed(am, thresh, mode='row'):
    am_thresh = threshold(am, thresh, mode=mode)
    am_norm = normalize(am_thresh)
    evals, evecs = np.linalg.eigh(am_norm)
    # Check if the matrix was singular (resulting in imaginary eigenvalues/eigenvectors).
    if np.iscomplex(evals).any() or np.iscomplex(evecs).any():
        raise RuntimeError('Complex evals/evecs; matrix might be singular or asymetric')
    return am_norm, am_thresh, np.flip(evals), np.flip(evecs)


'''
Compute the effective dimensionality of the input vector of eigenvalues.
'''
def effective_dimensionality(seq):
    seq = seq[1:] # Exclude the first eigenvalue.
    x1 = np.sum(np.abs(seq)) ** 2
    x2 = np.sum(seq ** 2)
    return (x1 / x2) / seq.shape[0]
```

```python
# read the mat file.
mat_contents = sio.loadmat('AllFigureData.mat')

# load the data for only the relevant states.
states = ['WA', 'S', 'U']
transition_mats = mat_contents['proEmbEx2'][0][0][0]
transition_mats = np.stack([i[0] for i in transition_mats])
transition_mats[np.isnan(transition_mats)] = 0

# plot the raw connectivity matrices.
fig, axes = plt.subplots(1, 3, figsize=(12,4), constrained_layout=True)
connectivity_matrices = {}
connectivity_matrices['raw'] = {}
connectivity_matrices['preprocessed'] = {}
for i, state in enumerate(states):
    raw_mat = transition_mats[i]
    mat, _, _, _ = embed(raw_mat.copy(), 0.5, mode='global')
    connectivity_matrices['raw'][state] = raw_mat / raw_mat.max()
    connectivity_matrices['preprocessed'][state] = mat
    cax = axes[i].imshow(connectivity_matrices['raw'][state], cmap='jet', interpolation='none')
    axes[i].set_title(state, fontsize=20)
    axes[i].set_xticks([])
    axes[i].set_yticks([])
fig.colorbar(cax, ax=axes[-1])
```

```python
# Compute the fraction of variance captured by a k-rank approximation.
states = ['WA', 'S', 'U']
fig, ax = plt.subplots(1,1,figsize=(5,5))
for state in states:
    matrix = connectivity_matrices['preprocessed'][state]
    s = np.linalg.svd(matrix)[1]**2
    s = np.linalg.eigh(matrix)[0][::-1]
    variances = s / (s).sum()
    plt.plot(np.cumsum(variances), label=state)
ax.set_xlim(0,75) 
plt.legend()
```

```python
# compute the SVD for each transition matrix and plot the singlar values.
states = ['WA', 'U']
all_singular_values = {}
fig, ax = plt.subplots(1,1,figsize=(5,5))
for i, state in enumerate(states):
    mat = connectivity_matrices['preprocessed'][state]
    singular_values = np.linalg.eigh(mat)[0][::-1]
    all_singular_values[state] = singular_values
    ax.plot(np.arange(len(transition_mats[i])), singular_values, label=states[i])
ax.legend()
#ax.set_ylim(-0.02,1)
ax.set_xlim(0,75) 
```
